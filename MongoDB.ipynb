4. How does MongoDB ensure high availability using replication ?
   - MongoDB ensures high availability through replica sets, which maintain multiple copies of data across servers. If the primary node fails, one of the secondary nodes is automatically promoted to primary, ensuring continuous service without downtime.
   - Replica Set
       - A group of mongod processes that store the same dataset.
   - Primary Node
       - Handles all write operations and replicates data to secondary nodes.
   - Automatic Failover
       - If the primary node goes down, the replica set automatically elects a new primary from the secondaries.
   - Arbiters
       - Optional nodes that participate in elections but don't store data, useful for maintaining odd numbers of voting members.

7. Describe the stages of the MongoDB, and how do they improve performance ?
   - In MongoDB, the "stages" usually refer to the steps in the aggregation pipeline. Each stage transforms the documents as they pass through, operations. This staged approach improves performance by reducing data early, optimizing memory usage, and leveraging indexes where possible.
   - Stages of MongDB Aggregation Pipeline
        - $match
             - Filters documents based on conditions
             - Performancee benefit : Reduces the dataset early, minimizing the amount of data processed in later stages.
             - Example : { $match : { status : "active" }}   
        - $project
             - Selects specific fields or reshapes documents.
             - Performance benefit : Limits unnecessary fields, reducing  memory and network overhead.
             - Example : { $project : { name : 1, age : 1 }}
  
        - $group
             - Groups documents by a field and performs aggregations.
             - Performance benefit : When combined with indexes, sorting is much faster and avoids in-memory sorting.
             - Example : { $sort : { age : -1 }}
        - $limit
             - Restricts the number of documents or skip a set.
             - Performance benefit : Prevents unnecessary processing of large datasets.
             - Example : { $limit : 10 }
               
        - $lookup
             - Performs a left outer join with another collection.
             - Performance benefit : Enables relational-style queries without duplicating data.
             - Example : {"$lookup" : {from : "orders", localField: "userId", foreignField :"uid", as : "orders" }}
               
        - $unwind
             - Deconstructs arrays into individual documents.
             - Performance benefit : Makes array data queryable and easier to aggregate.
             - Example : { $unwind : "$tags" }

    - How Stages Improve Performance
         - Early Filtering ($ match)
             - Cuts down data volume before heavy operations.
         - Field Reduction ($ project)
             - Minimizes memory and network usage.
         - Index Utilization
             - Stages like ($match) and ($sort) can leverage indexes for faster execution.
         - Parallelization
             - MongoDB can process stages stages efficiently  across shards in distributed setups.
         - Modularity
             - Breaking queries into stages makes them easier to optimize and debug.

       - Definition
            - sharding is the process of splitting a large dataset into smaller chunks and distributing them across multiple servers.

       - How it works:
           - A shard key determines how data is partitioned.
           - Each shard contain a subset of data.
           - A mongos router directs queries to the appropriate shard.
       - Performance Benefit
           - Queries only hit relevant shards, reducing load and improving response times for massive datasets.



    - Replication in MongoDB

       - Purpose

       - How it works
           - One primary node handles all writes.
           - Secondary nodes replicate data from the primary.
           - If the primary fails, a secondary is automatically promoted.
       - Performance Benefit
           - Improves read scalability and ensures data durability.

10. What are the ACID properties in the context of MongoDB transcations ?
    - Atomicity
        - All operations in a transaction are treated as single unit.
        - Either all changes are applied or none are applied.
        - Example : Transferring money between accounts - both debit and credit must succeed together, or neither happens.
        - MongoDB Benefit : Prevents partial updates that could corrupt data.
        
    - Consistency
        - Transactions move the database from one valid state to another.
        - Ensures that constraints, rules, and schemas  are respected.
        - Example : If a schema requires a unique email, a transaction won't allow duplicates.
        - MongoDB Benefit: Maintains data integrity across collections
    - Isolation
        - Transactions run independently of each other.
        - Intermediate states are hidden untill the transaction commits.
        - Example : Two users updating the same record won't see each other's partial changes.
        - MongoDB Benefit : Prevents race conditions and dirty reads in concurrent workloads.
    - Durability
        - Once a transaction is committed, changes are permanent - even if the system crashes.
        - MongoDB writes committed, changes are permanent even if the system crashes.
        - MongoDB writes committed data to disk and ensures recovery after failures.
        - MongoDB Benefit : Guarantees  reliability in production enviroments.

 - Workflow of a MongoDB Transaction
     - Active State                    : Transaction begins and executes read/write operations.

     - Partially Committed State       : Operations complete but are not yet permanent.

     - Committed State                 : Changes are saved successfully.

     - Aborted State                   : If an error occurs, all changes are rolled back.

     - Failed State                    : Transaction ends unsuccessfully

15. Explain the concept of embedded documents and when they should be used ?
    - Concept of Embedded Documents
        - Definition
            - An embedded document is a sub-document stored inside a parent document.
        - Structure
            - Uses nested curly braces {} to represent hierarchical data.
        - Analogy
            - Similar to storing JSON objects inside other JSON objects.
        - Limitations
            - MongoDB allows up to 100 levels of nesting, and the maximum document size is 16 M.

   - When to Use Embedded Documents
       - Data is closely related
           - Example: A user document containing an array of addresses or orders.
           - Since addresses belong only to that user, embedding avoids unnecessary references.
       - Frequent joint access
           - If you often query related data together,embedding ensures faster retrieval in a single query.
       - Hierarchical relationships
           - Example: Blog posts with embedded comments.
           - Comments are naturally tied to the post, so embedding reflects the hierarchy.
       - Performance optimization
           - Embedding reduces the need for cross-collection joins, which can be costly in NoSQL databases

19. How do MongoDB transaction differ from SQL transactions ?
    - Key Differences Between MongoDB and SQL Transactions
        - Data Model
            - SQL Transactions
                - Operate on rows and tables in a relational schema.
            - MongoDB Transactions
                - Operate on documents in collections, which can be nested and flexible in structure.
        - Multi-Document Support
            - SQL     -
                - Multi-row and multi-table transactions are native and fundamental.
            - MongoDB
                - Initially supported atomic operations only at the document level. Since version 4.0, MongoDB supports multi-document ACID transactions, making it closer to SQL in functionality.
        - ACID Properties
            - SQL
                - ACID compliance is built-in and expected.
            - MongoDB
                - Single-document operations are always atomic. Multi-document transactions are ACID-compliant but may introduce performance overhead compared to SQL systems optimized for them.
        - Performance Considerations
            - SQL
                - Optimized for transactional workloads (banking, inventory systems).
            - MongoDB
                - Optimized for high-volume, flexible workloads. Transactions are available but can reduce performance compared to single-document atomic operations.
        - Use Cases
            - SQL Transactions
                - Best for scenarios requiring strict consistency across multiple tables.
            - MongoDB Transactions
                 - Useful when multiple documents across collections must be updated consistently. However, MongoDB encourages schema design that minimizes the need for multi-document transactions by embedding related data.

23. What is MongoDB’s WiredTiger storage engine, and why is it important?
    - Document-Level Concurrency
        - Provides fine-grained locking at the document level.
        - Allows multiple clients to read and write different documents simultaneously.
        - Benefit: Improves throughput compared to older engines that used collection-level locks.
    - Compression
        - Supports data compression (Snappy, Zlib, Zstd).
        - Reduces storage footprint significantly.
        - Benefit: Lower disk usage and faster I/O operations.
   - ACID Transactions
       - WiredTiger supports multi-document ACID transactions (introduced in MongoDB 4.0).
       - Ensures atomicity, consistency, isolation, and durability across multiple operations.
       - Benefit: Makes MongoDB suitable for financial and enterprise-grade applications.
   - Checkpoints and Journaling
       - Uses write-ahead logging for durability.
       - Periodic checkpoints ensure recovery after crashes.
       - Benefit: Strong data integrity and fault tolerance.
   - Memory Efficiency
       - WiredTiger uses cache memory efficiently, balancing between RAM and disk.
       - Benefit: Handles large datasets with better performance.

9.Use aggregation to group orders by Region and calculate total sales per region.

```python
from pymongo import MongoClient
client = MongoClient("mongodb://localhost:27017/")
db = client["your_database_name"]
orders = db["Orders"]
pipeline = [
    {"$group": {"_id": "$Region", "totalSales": {"$sum": "$Sales"}}}
]
results = orders.aggregate(pipeline)
for result in results:
    print("Region:", result["_id"], "Total Sales:", result["totalSales"])
```
