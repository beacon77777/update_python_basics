6.How do we handle categorical variables in Machine learning? What are the coechiques?
  - Techniques for Handling Categorical Variables
    - Label Encoding
      - Assigns each category a unique integer.
      - Example : color = { Red: 0, Blue: 1, Green: 2 }
    
    - One-Hot Encoding
      - Creates binary columns for each category.
      - Example : color = Red, Blue, Green -> Red = [1,0,0], Blue = [0,1,0], Green = [0,0,1]

   - Dummy Encoding
     - Similar to one-hot encoding but drops one column to avoid redundancy.
     - Example : For 3 categoris, only 2 binary columns are created.

   - Count Encoding
     - Replace categories with their frequency or count in the dataset.
     - Example : If "Red" appears 50 times, "Blue" 30 times, "Green" 20 times, values become {Red : 50, Blue : 30, Green : 20}.
     Useful for high-cardinality categorical variables.

   - Target Encoding
     - Replace categories with the mean of the target variable for that category.
     - Example: If predicting sales, and "Region A" has average sales of 200, "Region B" has 150 -> encode accordingly.
     - Risk : Can cause data leakage if not applied carefully.

   - Binary Encoding
     - Converts categories into binary digits and splits them into separate columns
     - Example : Category "5" -> binary 101 -> three columns.
     - Efficient for high-cardinality features.

   - Embedding Representations
     - Used in deep learning . Categories are mapped to dense vactors in a continous space.
     - Example : Word embeddings for text categories.
     - Best for : Complex categorical data like text, user IDsm or product IDs.

12.Why do we have to perform EDA before fitting a model to the data?
   - Understand the Data Structure
     - EDA helps you see what variable exist, their types, and how they're distributed.
     - Without this, you might misapply algorithms categorical data as numeric.

   - Detect Missing Values and Outliers
     - Models don't handle missing values or extreme outliers well.
     - EDA shows where data cleaning or imputation is needed.

   - Spot Data Leakage or Bias
     - Sometimes target information sneaks into features.
     - EDA helps catch such issues early.

   - Feature Relationships
     - Visualizing correlations, scatterplots, and distributions reveals which features are most predictive.
     - Example : If two features are highly correlated, you might drop one to avoid redundancy.

   - Choose the Right Model and Preprocessing
     - EDA guides whether scaling, encoding, or transformations are required.
     - Example : If features are skewed, log-transform may help.

   - Prevent Garbage-in Garbage-out
     - a model is only as good as the data fed into it.
     - EDA ensures the dataset is trustworthy and meaningful before training.

21.What is feature scaling? How does it help in Machine Learning?
   - Feature scaling is the process of bringing all numerical features to a similar scale so that no single feature dominates the learning process just because of its magnitude.
   - Example
     - Feature 1 = Age (20-60 years)
     - Feature 2 = Salary(20,000 - 2,00,000)
     - Without scaling, salary values are much larger than age values, so the model may give more importance to salary - even if age is equally important.
  
  - Why is Feature Scaling Importing in Machine Learning.
    - Fairness among features
      - Prevents large scale feature from overpowering smaller ones.
    - Faster convergence
      - Gradient descent reaches the minimum more quickly when features are scaled.
    - Better accuracy
      - Models like KNN, SVM, PCA, and clustering rely on distances, scaling ensures correct distance calculations.
    - Stable optimization
      - Helps avoid exploding or vanishing gradients in neural networks.


22.How do we perform scaling in Python?
   - 1.Normalization (Min-Max Scaling)

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
data = np.array([[10], [20], [30], [40], [50]])
scaler = MinMaxScaler()
scaled = scaler.fit_transform(data)
print("Normalized data:\n", scaled)
```

2. Standardization (z-score Scaling)
   - Centers data around mean = 0 and standard deviation = 1

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(data)
print("Standardized data:\n", scaled)
```

3.Robust Scaling
  - Uses median and interquartile range, less sensitive to outliers.

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaled = scaler.fit_transform(data)
print("Robust scaled data:\n", scaled)
```

4.MaxAbs Scaling
  - Scales by maximum absolute value, keeping data in [-1,1]

```python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
scaled = scaler.fit_transform(data)
print("MaxAbs scaled data:\n", scaled)
```

24.How do we split data for model fitting in python?
   - To split data for model fitting in python, the most common approach  is to use scikit learn's train_test_split function. This ensures your model learns from one portion of the data and is evaluated on another,giving a fair estimate of performance.
   

```python
##Import the funcation
from sklearn.model_selection import train_test_split
import numpy as np



# Example features (X) and target (y)
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])
y = np.array([2, 4, 6, 8, 10, 12, 14, 16])




##Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

print("Training size:", len(X_train))
print("Testing size:", len(X_test))
```
