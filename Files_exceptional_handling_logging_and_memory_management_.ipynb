6.What is the difference between import and from ...import in python ?
  - Both import and from...import are used to bring external modules or parts of modules into your python program, but they differ in scope and usage style.
  
       -  Import:
            - Syntax : import module_name
            - Scope : import the entire module.you must reference functions/classes with the module prefix.
            - usage : math.sqrt(16)
            - Readability : makes it clear where a function/class comes from(good for large projects)
            - Namespace pollution : minimal - everything stays under the moidule's namespace.
            
      - from_import
                 
          - Syntax : from module_name import specific_name.
           - Scope : imports only specific functions, classes,or variables directly into your namespace.
           - Usage : from math import sqrt- sqrt(16)
          - Readability : Shorter code, but can reduce clarity if overused.
          - Namespace Pollution : Higher risk - directly adds names into your namespace, which can cause conflicts.

7.How can you handle multiple exceptions in python ?
  - In python, you can handle multiple exceptions by either:
     - Using multiple except blocks for different exception types.
     - Catching multiple exceptions in a single except block using a tuple.
     

```python
# Multiple except blocks
try:
  num = int(input("Enter the number: "))
  result = 10 / num
except ZeroDivisionError:
  print("Error : Division by Zero! ")
except ValueError:
  print("Error : Invalid input, Please enter a valid number.")
```

```python
# Single except block with tuple

try:
  num = int(input("Enter the number: "))
  result = 10 / num
except (ZeroDivisionError, ValueError) as e:
  print("Error:" , e)
```

9.What is the difference between multithreating and multiprocessing ?
  - Multithreading
    - Definition
       - Multiple threads run within the same process.

    - Shared memory
       - All threads share the same memory space.
    - Best for
       - Tasks that are input/output bound.
    - Limitation in Python
       - Due to the Global interpreter lock only one thread executes python bytecode at a time. so true paralleism is limited for CPU-heavy tasks.
    - Analogy
       - Imagine the same fridge and utensils.They can work together,but sometimes they bump into each other.


 - Multiprocessing
    - Definition
       - Multiple processes run independently, each with its own python interpreter and memory space.
    - No GIL Limitation
       - Each process runs in parallel on different CPU cores.
    - Best for
       - Tasks that are CPU-bound (heavy calculations,data crunching, simulations)
    - Analogy
       - Imagine multiple processes, each with its own fridge and utensils.The processes don't interfere with each other, but communication between kitchens is slower.

18.What is the difference between os.fork() and multiprocessing in Python ?
   - os.fork()
       - Platform support
            - Works only on Unix/Linux systems
       - Mechanism
            - Creates a child process by duplicating the parent's memory space.
       - Ease of Use
            - Low-level,required manual handling of process IDs,signals, and communication
       - Communication
            - No built-in support; must use OS-level IPC (pipes,socklets)
       - Safety
            - Can be error prone: child inherits parent's resources directly
       - Use Case
            - Suitable for simple process creation when you want exact duplication.

   - Multiprocessing
      - Platform Support
           - Works on both Unix/Linux systems
      - Mechanism
           - Spawns new processes using different start methods (fork,spawn,forkserver)
      - Ease of Use
           - High-level API with Process , Pool, Queue, Pipe.
      - Communication
           - Built-in support for inter-process communication and synchronization
      - Safety
           - Safer and more controlled, avoids issues with shared state.
      - Use Case
           - Best for parallelism,scaling tasks across multiple cores, and cross-platform relibility.

16.Demonstrate how to use memory profiling to check the memory usage of a small program.

```python
import tracemalloc
tracemalloc.start()
data = [i for i in range(1000000)]
current, peak = tracemalloc.get_traced_memory()
print(f"Current memory usage: {current / 1024**2:.2f} MB")
print(f"Peak memory usage: {peak / 1024**2:.2f} MB")
tracemalloc.stop()
```

18.How would you implement a basic logging setup that logs to a file with rotation after 1MB.

```python
import logging
from logging.handlers import RotatingFileHandler
handler = RotatingFileHandler("app.log", maxBytes=1_000_000, backupCount=3)
logging.basicConfig(handlers=[handler], level=logging.INFO)
logging.info("Program started")
try:
    x = 10 / 0
except ZeroDivisionError:
    logging.error("Division by zero error!")
logging.info("Program finished")
```

21.Write a Python program that reads a file and prints the number of occurrences of a specific word.

```python
filename = "example.txt"
word_to_count = "Python"
count = 0
with open(filename, "r") as f:
    for line in f:
        words = line.split()
        count += words.count(word_to_count)
print(f"The word '{word_to_count}' occurs {count} times in the file.")
```
