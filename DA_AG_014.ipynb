1.Explain the fundamental differences between DDL,DML, and DQL commands in SQL. Provide one example for each type of command.
  - Data Definition Language (DDL).
       - Purpose : Defines and manages the struture of the database objects
       - Key Actions : Create, alter, drop.
       - Effect : Changes the scheme itself, not the data inside it.

       - Example : CREATE TABLE Students (
                      StudentID INT PRIMARY KEY,
                      Name VARCHAR(50),
                      Age INT
                   );

  - Date Manipulation Language (DML)
      - Purpose : Deals with the actual data stored inside tables.
      - key Actions : Insert, Update, delete.
      - Effect : Modifies the records but does not change the table's structure.
      
      - Example :  INSERT INTO Students (StudentID, Name, Age)
                   VALUES (1, 'Rahul', 21);

  - Data Query Lanaguage (DQL)
      - Purpose : Focuses on retrieving data from the database.
      - Key Action : SELECT.
      - Effect : Does not modify data or structure, only fetches information.

      - Example : SELECT Name, Age
                  FROM Students
                  WHERE Age > 18;

4.What is a common table expression(CTE) in SQL, and what are its main benefits? Provide a simple SQL example demonstrating its usage.
  - A common table expression(CTE) is a temporary, named result set defined within the execution  scope of a single SQL statement.
  - It's created  using the WITH keyword.
  - Think of it as a 'mini-table' or a reusable subquery that exists only while your query runs.

Main Benefits of CTEs

  - Readability and Organization
      - Breaks complex queries into smaller, logical chunks
      - Makes SQL easier to understand and maintain.
  
  - Reusability
      - You can reference the same CTE multiple times in a query, avoiding duplicate subqueries.
  - Recursive Queries
      - CTEs support recursion, which is useful for hierarchical data


      - WITH AdultStudents AS(

           )

  - Normalization is the process of organizing data in a database to reduce redundancy and improve data integrity.
  - It divides largee, complex tables into smaller, related ones.
  - Ensures data is stored logically and consistenly.

First Normal Form (1NF)
  - Rule : Each column must hold atomic values, and each row must be unique.
  - Scenario : A Students table should not have a column like Phonenumbers storint multiple numbers in one cell. Instead, each phone number should be in a separate row.

Second Normal Forms (2NF)
  - Rule : Must be in 1NF, and all non-key attributes must be depend on the entire primary key, not just part of it.
  - Scenario : In a table with composite key (StudenID, CourseID), storing StudentName would violate 2NF because StudentName depends only on StudentID, not on the combination.

Third Normal Form (3NF)
   - Rule : Must be in 2NF, and all non-key attributes must depends only on the primary key, not on other non-key attributes
   - Scenario : If a Students table has StudentID, DepartmentID, DepartmentName, StudentID. This violate 3NF.

9.Write a SQL query that uses a Common Table Expression (CTE) and a Window Funcation (specifically ROW_NUMBER(9) or RANK()) to display the CategoryName,ProductName, ProductName, and Price for the top 2 most expensive products in each CategoryName.


```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

cursor.execute('PRAGMA foreign_keys = ON;')
cursor.execute("""
CREATE TABLE Categories (
    CategoryID INT PRIMARY KEY,
    CategoryName VARCHAR(50) NOT NULL UNIQUE
);
""")
cursor.execute("""
CREATE TABLE Products (
    ProductID INT PRIMARY KEY,
    ProductName VARCHAR(100) NOT NULL UNIQUE,
    CategoryID INT,
    Price DECIMAL(10,2) NOT NULL,
    StockQuantity INT,
    FOREIGN KEY (CategoryID) REFERENCES Categories(CategoryID)
);
""")

# Insert data into Categories
categories_data = [
    (1, 'Electronics'),
    (2, 'Books'),
    (3, 'Home Goods'),
    (4, 'Apparel')
]
cursor.executemany("INSERT INTO Categories (CategoryID, CategoryName) VALUES (?, ?)", categories_data)

# Insert data into Products (corrected and formatted)
products_data = [
    (101, 'Laptop Pro', 1, 1200.00, 50),
    (102, 'SQL Handbook', 2, 45.50, 200),
    (103, 'Smart Speaker', 1, 99.99, 150),
    (104, 'Coffee Maker', 3, 75.00, 80),
    (105, 'Novel: The Great SQL', 2, 25.00, 120),
    (106, 'Blender X', 3, 120.00, 60),
    (107, 'Wireless Earbuds', 1, 150.00, 100),
    (108, 'T-Shirt Casual', 4, 20.00, 300)
]
cursor.executemany("INSERT INTO Products (ProductID, ProductName, CategoryID, Price, StockQuantity) VALUES (?, ?, ?, ?, ?)", products_data)

# Commit the changes
conn.commit()

# SQL query with CTE and Window Function
sql_query = """
WITH RankedProducts AS (
    SELECT
        c.CategoryName,
        p.ProductName,
        p.Price,
        ROW_NUMBER() OVER (
            PARTITION BY c.CategoryName
            ORDER BY p.Price DESC
        ) AS RowNum
    FROM Products p
    JOIN Categories c
        ON p.CategoryID = c.CategoryID
)
SELECT
    CategoryName,
    ProductName,
    Price
FROM RankedProducts
WHERE RowNum <= 2
ORDER BY CategoryName, Price DESC;
"""
df_result = pd.read_sql_query(sql_query, conn)
print(df_result)
conn.close()
```
