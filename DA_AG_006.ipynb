1. What is a random variable in probability theory ?
   - In probability theory, a random variable is a mathmatical funcation that assigns numerical values to the outcomes of a random experiment. instead of dealing directly with abstract outcomes, we use random variable to represent them as numbers, making analysis easier.

   Definition : A random variable is a function X : S -> R, where:
   
   - S = sample space
   - R = set of real numbers

   Example:

    - Tossing a die -> sample space = { 1,2,3,4,5,6}
    - Random variable X could assign each outcome directly as its face value.

   Types:
   
    - Discreate random variable : Takes countable values.
    - Continuous random variable : Takes values from an interval of real numbers.

7.What is the  significance of intervals in statistical analysis ?
  - A confidence interval (CI) is an estimated range of values, calculated from sample data, that is likely to contain the true population parameter with a certain level of confidence.

  - if the average height of a sample of students is 170 cm with a 95% CI of [168,172], we can say we are 95% confident the true average heigh of the population lies between 168 and 172 cm.


1.Account for Uncertainty
  - Unlike a single point estimate, Cls acknowledge sampling variability and give a realistic range.

2.Supports Decision-Making
  - Widely used in research, medicine,economics, and engineering to assess reliability of estimates.
  - Example: In clinical trials, Cls help determine whether a treatment effect is statistically meaningful.

3.Connects to Hypothesis Testing
  - If a CI for a mean difference does not include 0, It suggests a statistically significant difference.
  - This makes Cls a more intuitive alternative to P-values.

4.Precision indicator
  - Narrow intervals : more precise estimates.
  - Wide intervals   : less precise, more uncertainty.

9.Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5 . Compute its mean and standard deviation using NumPy, and draw a histoigram to visualize the distribution.
  -

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc = 50, scale = 5, size = 1000)

mean_value = np.mean(data)
std_value = np.std(data)

print(f"Computed Mean:", mean_value)
print(f"Computed Standard Deviation: ", std_value)

plt.hist(data, bins = 30, color = 'skyblue', edgecolor = 'black')
plt.title('Histogram of Normal Distribution (Î¼=50, Ïƒ=5)')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(mean_value, color = "red", linestyle = "dashed", linewidth = 2, label = f"Mean" = {mean_value:.2f}")
plt.legend()
plt.show()
```
