5. What is mutability in Python?
   - Mutability refers to whetherr the contents of an object can be changed after creation.

      - Mutable objects -> can be modified
      - Immutable objects -> can not be changed once created.


```python
# mutable objects (list)
marks = [10,20,15,45,12]
marks[0]=50
print(marks)
```

```python
# Immutable objects (tuple)
marks = (10,20,15,45,12)
marks[2]=50
print(marks)
```

6. Why are lists mutable,but tuples are immutable?
   - list -> Python stores lists as dynamic arrays. This means elements can be added, removed, changed without creating a new object.
   
   - Tuples -> Once created, the memory layout of a tuple is fixed. Python does not allow modification of its elements.
   

```python
## list example
fruits = [ 'Apple','Banana','Dragon Fruits','Orange']
fruits[1] = 'kiwi'
print(fruits)
```

```python
## Tuples example
fruits = ( 'Apple','Banana','Dragon Fruits','Orange')
fruits[1] = 'kiwi'
print(fruits)
```

10. Write code to demonstrate type casting with list elements.

```python
# demonstrate type casting with list elements
data = ['10','21','25','24']
print("Original list : ",data)

#convert to integers
int_list = [int(x) for x in data if x.isdigit()]
print("Integers:", int_list)

#convert to floats
float_list = [ float(x) for x in data if x.replace('.','',1).isdigit()]
print("Floats : ",float_list)

#convert to booleans
bool_list = [bool(x) for x in data]
print("Booleans : ",bool_list)
```

14. Write a program to reverse a string using a while loop.

```python
name = input("Enter your name : ")
reverse = ""
i = len(name) - 1
while i >= 0:
  reverse += name[i]
  i -= 1
  print("Reversed name : ",reverse)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
# Program to calculate factoria using while loop
num = int(input("Enter a number : "))
factorial = 1
i = 1
while i <= num:
  factorial *= i
  i += 1
print("Factorial of ", num , "is",factorial)
```
