20.Explain the difference between a list and a dictionary in terms of data retrieval.
   - List : Index-Based Retrieval
        - lists store elements in an ordered sequence.
        - Data is retrieved using a numberic index (position)
        - You must Know the position of the element.

   - Dictionary : Key-Based Retrieval
        - Dictionaries store data as Key-value pairs.
        - Data is retrived using a Unique key, not position.
        - Keys are meaningful and descriptive.

```python
# Examples  - List (Index-Based Retrieval)
fruites = ["apple","banana","orange"]
print(fruites[1])
```

```python
# Examples - Dictionary(Key-based Retrieval)
dict1= {"fruit1":"apple","fruit2":"kiwi","fruit3":"orange"}
print(dict1["fruit2"])
```

    Feature	                   List	                             Dictionary
    Access Method	        Index (0,1,2…)	                     Key
    Retrieval Speed	        Fast by index, slow by value	     Very fast by key
    Data Structure	        Ordered collection	                 Key–value mapping
    Use Case	            Sequential data	                     Labeled/structured data
