2.What is sampling in statistics ? Explain the differences between random and stratified sampling.
  - Random Sampling

     - Definition : Every individual in the population has an equal
                  chance of being selected.
     - Method    : Use random number generator, lottery method, or
                  Simple random draws.
     - Example  : Suppose you want to study the average height of
                  Students in a university.
                  You randomly select 100 students from the entire student list, without considering their department, year or gender.

  - Stratified Sampling

     - Definition   :  The population is divided into strata based on
                       specific characteristics. Then random samples are
                       taken from each stratum proportionally.
     - Method       :  Identify strate - decide sample size for each
                       Randomly select within each stratum.
     - Example      : In the same university, students are divided into
                      Department. If science has 50% of students, Arts 30%, and Commerce 20%, you select your sample in the same proportion.




3.Define mean, median, and mode. Explain why these measures of central tendency are important.
  - Mean (Average)
      - The mean is the sum of all
                                                  Values in a dataset divided by
                                                  The number of values.

                 Mean =  (Sum of all values) / Number of values
Examples :  For data 2,4,6 the mean is   (2+4+6)/3   =  4

  - Median
     - The median is the middle value when the data is
                 Arranged in ascending or descending order.
     - Example : for data [2,4,6], median = 4.
                 for [2,4,6,8], median = (4+6)/2 = 5
 - Mode
     - The mode is the value that occurs most frequently in the dataset.
     - Example : for data [2,4,4,6], mode = 4


4.Explain skewness and kurtosis. What does a postive skew imply about the data?
  - Skewness
      - Definition
         -  measures the asymmetry of a data distribution around its mean.
     - Types
         - Zeros Skewness
             - Perfectly symmetrical distribution.
         - Postive skew
             - The tail on the right side is longer or fatter than the left.
         - Negative skew
             - The tail on the left side is longer or fatter than the right.

  - Kurtosis
      - Definition
           - Kurtosis measures the "tailedness" or the sharpness of the peak of a distribution compared to a normal distribution.

      - Types
           - Mesokurtic
              - Normal distribution
           - Leptokurtic
              - Sharper peak, heavier tails
           - Platykurtic
              - Flatter peak, lighter tails.

5. Implement a Python program to compute the mean, median, and mode of a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28].
   

```python
import statistics

num = [12,15,12,18,19,12,20,22,19,19,24,24,26,28]
mean_value = statistics.mean(num)
median_value = statistics.median(num)
mode_value = statistics.mode(num)
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

6.Compute the convariance and correlation coefficient between the following two datasets as lists in python ?

list_x = [10,20,30,40,50]

list_y = [15,25,35,45,60]


```python
import numpy as np
list_x = [10,20,30,40,50]
list_y = [15,25,35,45,60]

x = np.array(list_x)
y = np.array(list_y)

cov_value = np.cov(x, y, bias=True)
cov_xy = cov_value[0, 1]
corr_xy = np.corrcoef(x,y)[0,1]

print("Covariance:", cov_xy)
print("Correlation Coefficient : ", corr_xy)
```

7.Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result.
  data = [ 12,14,14,15,18,19,21,22,22,23,23,24,26,29,35 ]
  

```python
import matplotlib.pyplot as plt
import numpy as np

data = [12,14,14,15,18,19,21,22,22,23,23,24,26,29,35]
plt.boxplot(data, vert=False, patch_artist=True, boxprops = dict(facecolor = "lightblue"))
plt.title("Boxplot of Data with Outliers")
plt.xlabel("Values")
plt.show()

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1",Q1)
print("Q3",Q3)
print("IQR",IQR)
print("Lower Bound",lower_bound)
print("Upper Bound",upper_bound)
print("Outliers:", outliers)
```

8.You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
  - Explain how you would use covariance and correlation to explore this relationship.
  - Write Python code to compute the correlation between the two lists.

advertising_spend = ( 200,250,300,400,500 )
daily_sales = ( 2200,2450,2750,3200,4000 )


```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

x = np.array(advertising_spend)
y = np.array(daily_sales)

corr_xy = np.corrcoef(x,y)[0,1]

print("Correlation Coefficient : ", corr_xy)
```

9. Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
   - Explain which summary statistics and visualization you had use.
   - Write Python code to create a histogram using Matplotlib for the survey data:

   - survey_scores = ( 7,8,5,9,10,4,7,6,9,8,7 )

```python
import matplotlib.pyplot as plt

survey_scores = [7,8,5,9,10,4,7,6,9,8,7]

plt.hist(survey_scores, bins = 6, color = "skyblue", edgecolor = "black")
plt.title("Customer Satisfaction Survey Distribution")
plt.xlabel("Satisfaction Score (1-10)")
plt.ylabel("Frequency")
plt.show()
```
