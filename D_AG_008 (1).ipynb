2.What are the key assumptions of Simple Linear Regression ?
  - Linearity
    - The relationship between the independent variable (X) and dependent variable (Y) is linear.
    - This means changes in X lead to proportional changes in Y.

  - Independence of Errors
    - The residuals are independent of each other.
    - No autocorrelation should exist.

 - Homoscedasticity
   - The variance of residuals is constant across all values of X.
   - In other words, the spread of errors should not increase or decrease with X.

 - Normality of Errors
   - The residuals should be approximately normally distributed.
   - This is especially important for hypothesis testing and confidence intervals.

 - No Perfect Multicollinearity
   - Since SLR involves only one predictor, this assumption is trivial here.
   - But in multiple regression, predictors should not be perfectly correlated.

 - Measurement Accuracy
   - X and Y should be measured accurately without systematic errors.
   - Measurement errors in X can bias the slope estimate.

    - The value of 𝑌 when 𝑋 = 0.It represents the baseline level of the dependent variable.

  - β1 (Slope Coefficient)
    - The change in 𝑌 for a one-unit increase in 𝑋. It shows the strength and direction of the relationship.
    - Example: Each extra hour of study increases the score by 5 points.

  - 𝜖 (Error Term):  
    - The difference between the observed value and the predicted value. It captures random variation and factors not explained by 𝑋

7. Name and briefly describe three common evaluation metrics for regression models.
   - When assessing how well a regression model performs, we often use these metrics:
     - 1.Mean Absolute Error
       - Formula:
        - 𝑀𝐴𝐸=1𝑛∑𝑖=1𝑛∣𝑌𝑖−𝑌^𝑖∣   
        - Meaning: Average of the absolute differences between actual values (𝑌𝑖) and predicted values (𝑌^𝑖).
        - Interpretation: Easy to understand since it represents the average error in the same units as the data.

    
     - 2.Mean Squared Error (MSE)

       - Formula:
         - 𝑀𝑆𝐸=1𝑛∑𝑖= 1𝑛(𝑌𝑖−𝑌^𝑖)2

       - Meaning: Average of squared differences between actual and predicted values.
       - Interpretation: Penalizes larger errors more heavily than MAE, making it sensitive to outliers.


     - 3.R-squared (𝑅2)
       -  Coefficient of Determination
          - Formula
            - 𝑅2 = 1−𝑆𝑆𝑟𝑒𝑠𝑆𝑆𝑡𝑜𝑡 where 𝑆𝑆𝑟𝑒𝑠 = sum of squared residuals, and 𝑆𝑆𝑡𝑜𝑡 = total sum of squares.
            - Meaning: Proportion of variance in the dependent variable explained by the model.
            - Interpretation: Ranges from 0 to 1. Higher values indicate a better fit.


8.What is the purpose of the R-squared metric in regression analysis ?
  - R-squared (𝑅2), also called the coefficient of determination, is a statistical measure that indicates how well a regression model explains the variability of the dependent variable.


  - What it Represents
    - 𝑅2 measures the proportion of variance in the dependent variable (Y) that is explained by the independent variable(X).
    - It ranges between 0 and 1.
      - 0 → The model explains none of the variability.
      - 1 → The model explains all the variability.


9.Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.
  -

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for sklearn
y = np.array([40, 50, 60, 70, 80])

model = LinearRegression()
model.fit(X, y)

print("Slope (β1):", model.coef_[0])
print("Intercept (β0):", model.intercept_)
```

10.How do you interpret the coefficients in a simple linear regression model?
   - How to Interpret the Coefficients
     - Intercept (β₀)
       - It’s the predicted value of Y when X = 0.
       - Think of it as the “starting point” of the line.

     - Slope (β₁)
       - It tells you the rate of change.
       - For every 1‑unit increase in X, Y changes by β₁ units.
       - Positive slope → Y increases as X increases.
       - Negative slope → Y decreases as X increases.
