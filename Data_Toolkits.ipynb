2.How does broadcasting work in NumPy?
     
  - Broadcasting is NumPy's way of performing arithmetic opertions on arrays of different shapes without explicitly copying data.

  - idea : if two arrays don't have the same shape, NumPy automatically "stretches" the smaller one across the larger one so the operation makes sense.

  - Analogy : Imagine you're serving food at a party. if you have one plate of snacks and 5 guests, you don't need to prepare 5 identical plates - you just share the same plate across all guests.




```python
import numpy as np
aro = np.array([1, 2, 3])
result = aro + 5
print(result)
```

4.Explain the use of the groupby() method in Pandas.
  - The groupby() method in Pandas is used to split data into group based on some criteria, then apply operations on each group separately.

```python
# Basic grouping
import pandas as pd

data = {
"students" : ["Rahul","Anita","Batra"],
"age" : [25,30,35],
"city" : ["Lucknow","Ranchi","Mumbai"],
"Marks" : [80,90,95]
         }

df = pd.DataFrame(data)
result = df.groupby("students")["Marks"].mean()
print(result)
```

```python
# Multiple Aggregations

result = df.groupby("students")["Marks"].agg(["mean","max"])
print(result)
```

```python
# Grouping by Multiple columns
import pandas as pd
data = {
    "city" : [ "Lucknow","Delhi", "Mumbai", "Kolkata" ],
    "subjects" : ["math", "science", "social science", "biology"],
    "score" : [ 55,36,85,54]
}

df = pd.DataFrame(data)
result = df.groupby(["city","subjects",])["score"].mean()
print(result)
```

7. What is a heatmap, and when should it be used ?
  - A heatmap displays data values as colors, where darker shades indicate higher or lower values.
  - It's essentially a colored grid : rows and columns represent categories, and the cell color represents the magnitude of the data.

When should it be used ?
  - Comparing values across two dimension
  - Identifying patterns or trends
  - Highlighting correlations

```python
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
       "math" : [ 80,85,90,70],
       "science" : [ 78,88,92,65],
       "English" : [ 82,78,85,75]
})

corr = data.corr()
sns.heatmap(corr, annot = True, cmap = "coolwarm")
```

8. What does the term "vectorized operation" mean in NumPy ?
   - A vectorized operation in NumPy mean appling an operation directly to entire arrays without writing explicit loops.
   - Instead of processing elements one by one in Python, NumPy uses optimized C - based implementations under the hood to perform operations on all elements simultaneously.




```python
# Scalar Multiplication

import numpy as np
arr = np.array([1,2,3,4,5])
print(arr * 2)
result = [x * 2 for x in arr]
print(result)
```

```python
# Element wise Addition

arr1 = np.array([1,2,3,4,5])
arr2 = np.array([6,7,8,9,10])
print(arr1 + arr2)
```

```python
# Mathematical Funcations
arr = np.array([ 0, np.pi/2])
print(np.sin(arr))
```

17. Explain the difference between apply() and map() in Pandas.
    - Map()
       - Works on
           - Series only
       - Function applied
           - Element wise
       - Use case
           - Simple transformations
       - Axis parameter
           - Not available

    - apply()
       - works on
          - Series & DataFrame
       - Function applied
          - Element wise or row/column wise (DataFrame)
       - Use case
          - Complex operations across row/columns
       - Axis parameter
          - Available ( axis = 0 for columns, axis = 1 for rows )

```python
# Map()
import pandas as pd
s = pd.Series([1,2,3,4])
print(s.map(lambda x : x**2))

# apply()
s = pd.Series([1,2,3,4])
print(s.apply(lambda x : x ** 2 ))
```

20. What is the role of a pivot tables in Pandas ?
    - A pivot table in Pandas is used to summarized and reorganzie data
    - It helps you group, aggregate, and analyze large datasets quickly.
    - you can rearrange rows and columns to see data from different perspectives.



```python
import pandas as pd

data = {
    "city" : ["Lucknow","Delhi","Mumbai","Kolkata"],
    "Product" : ["A","B","A","B"],
    "Sales" : [100,200,150,300]
}

df = pd.DataFrame(data)
pivot_table = pd.pivot_table(df,index = "city", columns = "Product", values = "Sales", aggfunc = "sum")
print(pivot_table)
```
